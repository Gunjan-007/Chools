### Graphs

```python

# Load the mpg dataset from Seaborn
mpg_df = sns.load_dataset("mpg")

# Scatter plot of horsepower vs. mpg
plt.figure(figsize=(10, 6))
sns.scatterplot(x="horsepower", y="mpg", data=mpg_df, hue="origin", palette="viridis", size="weight", sizes=(20, 200))
plt.title("Scatter plot of Horsepower vs. MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.legend(title="Origin")
plt.show()
```

```python

# Box plot of mpg by number of cylinders
plt.figure(figsize=(10, 6))
sns.boxplot(x="cylinders", y="mpg", data=mpg_df, palette="Set3")
plt.title("Box plot of MPG by Number of Cylinders")
plt.xlabel("Number of Cylinders")
plt.ylabel("MPG")
plt.show()
```

```python

# Distribution of MPG
plt.figure(figsize=(10, 6))
sns.histplot(mpg_df["mpg"], bins=20, kde=True, color="skyblue")
plt.title("Distribution of MPG")
plt.xlabel("MPG")
plt.ylabel("Frequency")
plt.show()
```
