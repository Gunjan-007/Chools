### Tuning of Hyperparameters: Learning rate and Drop out rate

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam
from keras.wrappers.scikit_learn import BaseWrapper
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, mean_squared_error
import numpy as np

# Define a custom Keras wrapper class
class CustomKerasRegressor(BaseWrapper):
    def __init__(self, build_fn=None, **sk_params):
        super().__init__(build_fn, **sk_params)

    def build_fn(self, learning_rate=0.01, dropout_rate=0.0):
        return self.sk_params['build_fn'](learning_rate=learning_rate, dropout_rate=dropout_rate)

# Define the model
def create_model(learning_rate=0.01, dropout_rate=0.0):
    model = Sequential()
    model.add(Dense(13, input_dim=13, activation='relu'))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1))
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model

# Create the custom Keras regressor
model = CustomKerasRegressor(build_fn=create_model)

# Define the grid search parameters
param_grid = {
    'learning_rate': [0.0001, 0.001, 0.01],  # Reduced learning rate
    'dropout_rate': [0.0, 0.1, 0.2]
}

# Define the scoring function
scoring = make_scorer(mean_squared_error, greater_is_better=False)

# Perform grid search
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3, scoring=scoring)
grid_result = grid.fit(X, Y)

# Print the results
print("Best score: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
# Importing necessary libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam


# Function to build the model
def build_model(batch_size=32, epochs=100, learning_rate=0.001, dropout_rate=0.0, activation='relu', kernel_initializer='glorot_uniform', neurons=64):
    model = Sequential()
    model.add(Dense(neurons, activation=activation, kernel_initializer=kernel_initializer, input_dim=X_train_scaled.shape[1]))
    model.add(Dropout(dropout_rate))
    model.add(Dense(1, activation='linear'))

    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='mean_squared_error', optimizer=optimizer)
    return model

# Tuning of Hyperparameters: Batch Size and Epochs
def tune_batch_epochs():
    batch_sizes = [16, 32, 64]
    epochs_list = [50, 100, 150]

    for batch_size in batch_sizes:
        for epochs in epochs_list:
            model = build_model(batch_size=batch_size, epochs=epochs)
            model.fit(X_train_scaled, Y_train, batch_size=batch_size, epochs=epochs, verbose=0)
            mse = model.evaluate(X_test_scaled, Y_test, verbose=0)
            print(f'Batch Size: {batch_size}, Epochs: {epochs}, MSE: {mse}')

# Tuning of Hyperparameters: Learning Rate and Dropout Rate
def tune_learning_dropout():
    learning_rates = [0.001, 0.01, 0.1]
    dropout_rates = [0.0, 0.1, 0.2]

    for lr in learning_rates:
        for dropout_rate in dropout_rates:
            model = build_model(learning_rate=lr, dropout_rate=dropout_rate)
            model.fit(X_train_scaled, Y_train, batch_size=32, epochs=100, verbose=0)
            mse = model.evaluate(X_test_scaled, Y_test, verbose=0)
            print(f'Learning Rate: {lr}, Dropout Rate: {dropout_rate}, MSE: {mse}')

# Tuning of Hyperparameters: Activation Function and Kernel Initializer
def tune_activation_kernel():
    activations = ['relu', 'sigmoid']
    kernel_initializers = ['glorot_uniform', 'he_uniform']

    for activation in activations:
        for kernel_initializer in kernel_initializers:
            model = build_model(activation=activation, kernel_initializer=kernel_initializer)
            model.fit(X_train_scaled, Y_train, batch_size=32, epochs=100, verbose=0)
            mse = model.evaluate(X_test_scaled, Y_test, verbose=0)
            print(f'Activation: {activation}, Kernel Initializer: {kernel_initializer}, MSE: {mse}')

# Tuning of Hyperparameter: Number of Neurons in Activation Layer
def tune_neurons():
    neurons_list = [32, 64, 128]

    for neurons in neurons_list:
        model = build_model(neurons=neurons)
        model.fit(X_train_scaled, Y_train, batch_size=32, epochs=100, verbose=0)
        mse = model.evaluate(X_test_scaled, Y_test, verbose=0)
        print(f'Neurons: {neurons}, MSE: {mse}')

# Training model with optimum values of Hyperparameters
def train_with_optimal_hyperparameters():
    # Optimal values found from tuning
    model = build_model(batch_size=32, epochs=100, learning_rate=0.001, dropout_rate=0.2, activation='relu', kernel_initializer='glorot_uniform', neurons=64)
    model.fit(X_train_scaled, Y_train, batch_size=32, epochs=100, verbose=1)
    mse = model.evaluate(X_test_scaled, Y_test, verbose=0)
    print(f'MSE with Optimal Hyperparameters: {mse}')

# Hyperparameters all at once
def train_with_all_hyperparameters():
    model = build_model(batch_size=32, epochs=100, learning_rate=0.001, dropout_rate=0.2, activation='relu', kernel_initializer='glorot_uniform', neurons=64)
    model.fit(X_train_scaled, Y_train, batch_size=32, epochs=100, verbose=1)
    mse = model.evaluate(X_test_scaled, Y_test, verbose=0)
    print(f'MSE with All Hyperparameters: {mse}')

# Running each section
print("Tuning of Hyperparameters: Batch Size and Epochs")
tune_batch_epochs()

print("\nTuning of Hyperparameters: Learning Rate and Dropout Rate")
tune_learning_dropout()

print("\nTuning of Hyperparameters: Activation Function and Kernel Initializer")
tune_activation_kernel()

print("\nTuning of Hyperparameter: Number of Neurons in Activation Layer")
tune_neurons()

print("\nTraining model with optimum values of Hyperparameters")
train_with_optimal_hyperparameters()

print("\nTraining model with all hyperparameters")
train_with_all_hyperparameters()
```
